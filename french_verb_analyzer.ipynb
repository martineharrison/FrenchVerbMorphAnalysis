```python
import pynini
from pynini.lib import paradigms, features, pynutil, rewrite

def write_output(noun: str, pd: paradigms.Paradigm, filepath: file) -> None:
    lattice = rewrite.rewrite_lattice(
        noun,
        pd.stems_to_forms @ pd.feature_label_rewriter
    )
    for wordform in rewrite.lattice_to_strings(lattice):
        print(wordform, file = filepath)
    return

def print_forms(noun: str, pd: paradigms.Paradigm) -> None:
    lattice = rewrite.rewrite_lattice(
        noun,
        pd.stems_to_forms @ pd.feature_label_rewriter
    )
    for wordform in rewrite.lattice_to_strings(lattice):
        print(wordform)
    return
```
